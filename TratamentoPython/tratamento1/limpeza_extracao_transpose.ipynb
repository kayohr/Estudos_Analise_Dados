```python
# Substituir NaN e NaT por "EXCLUDE"
df.fillna('EXCLUDE', inplace=True)

# Substituir "EXCLUDE" por uma string vazia ('')
df.replace('EXCLUDE', '', inplace=True)

```

```python
import numpy as np

# Substituir strings vazias por NaN
df['Total'].replace('', np.nan, inplace=True)

# Converter a coluna "Total" para números de ponto flutuante e somar
total_sum = df["Total"].astype(float).sum()

print(total_sum)

```
